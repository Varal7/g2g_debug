```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Load Test Sests:

```python
eval_path='checkpoints/aqsol/eval'
```

```python
test_sets = ['bottom','med_bottom','medium','med_top','top']
```

```python
stats = {}
for test_set in test_sets:
    stats[test_set] = pickle.load( open(os.path.join(eval_path,'stats_'+test_set+'.pkl'), "rb" ) )
```

### Available Metrics:

```python
stats['bottom'].keys()
```

### Plot Metrics:

```python
for metric in ['avg_improvment','percent_improved','percent_improved_mae','translation_failure','avg_tanimoto','percent_above_sa']:
    print(metric)
    plt.plot([stats[x][metric] for x in list(stats.keys())],marker='o')
    plt.grid()
    plt.show()
```
